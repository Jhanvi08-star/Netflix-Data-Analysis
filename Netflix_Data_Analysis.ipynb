```python
## 1. Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

### Style settings
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

### Create images folder if not exists
os.makedirs("images", exist_ok=True)
```

```python
## 2. Loading the Dataset
df = pd.read_csv("/content/netflix_titles.csv")
df
```

```python
## 3. Dataset Overview
```

```python
### Show first 5 rows
df.head()
```

```python
### Show last 5 rows
df.tail()
```

```python
### Shape of dataset
print("Shape of dataset:", df.shape)
```

```python
### Column names
print("\nCoulmns in dataset:")
print(df.columns.tolist())
```

```python
### Dataset info
print("Dataset Info:")
df.info()
```

```python
### Check missing values
print("Missing Values:\n")
print(df.isnull().sum())
```

```python
### Basic statistics for numerical columns
df.describe(include='all')
```

```python
## 4. Data Cleaning

### Fill missing values in important columns
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unknown')
df['duration'] = df['duration'].fillna('Unknown')
```

```python
### Convert date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
### Drop rows where date_added is missing
df = df.dropna(subset=['date_added'])
```

```python
### Extract year and month added
df.loc[:, 'year_added'] = df['date_added'].dt.year
df.loc[:, 'month_added'] = df['date_added'].dt.month_name()
```

```python
### Check cleaned data
print("Missing values after cleaning:\n")
print(df.isnull().sum())
```

```python
## 5. Exploratory Data Analysis (EDA)
```

```python
### Check duplicate rows
print("Duplicate rows:", df.duplicated().sum())
```

```python
### Count of Movies and TV Shows
type_counts = df['type'].value_counts()
print(type_counts)
```

```python
### Plot

### Movies vs TV Shows on Netflix
sns.countplot(data=df, x='type')
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("images/movies_vs_tvshows.png")
plt.show()
```

```python
### Pie chart for type distribution

### Percentage of Movies vs TV Shows
type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Percentage Distribution of Movies and TV Shows")
plt.ylabel("")
plt.tight_layout()
plt.savefig("images/type_distribution_pie.png")
plt.show()
```

```python
### Top 10 Countries with Most Content

### Some rows have multiple countries, so explode them properly
country_data = df['country'].str.split(', ', expand=False).explode()

top_countries = country_data.value_counts().head(10)
print(top_countries)

### Plot
top_countries.plot(kind='bar')
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("images/top_countries.png")
plt.show()
```

```python
### Top 10 ratings
top_ratings = df['rating'].value_counts().head(10)
print(top_ratings)

### Plot
sns.countplot(data=df, y='rating', order=df['rating'].value_counts().index[:10])
plt.title("Top 10 Content Ratings on Netflix")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.tight_layout()
plt.savefig("images/top_ratings.png")
plt.show()
```

```python
### Count content added each year
year_added_counts = df['year_added'].value_counts().sort_index()
print(year_added_counts)

### Plot
year_added_counts.plot(kind='line', marker='o')
plt.title("Netflix Content Added Over the Years")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.tight_layout()
plt.savefig("images/content_added_by_year.png")
plt.show()
```

```python
### Content Added by Month

### Month order for proper sorting
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

month_counts = df['month_added'].value_counts().reindex(month_order)
print(month_counts)

### Plot
month_counts.plot(kind='bar')
plt.title("Netflix Content Added by Month")
plt.xlabel("Month")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("images/content_added_by_month.png")
plt.show()
```

```python
### Number of titles by release year
release_year_counts = df['release_year'].value_counts().sort_index()

### Plot only recent years for better readability
release_year_counts.tail(30).plot(kind='line', marker='o')
plt.title("Netflix Titles Released Over the Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.tight_layout()
plt.savefig("images/release_year_trend.png")
plt.show()
```

```python
### Top 10 Genres

### Split genres properly
genres = df['listed_in'].str.split(', ', expand=False).explode()

top_genres = genres.value_counts().head(10)
print(top_genres)

### Plot
top_genres.plot(kind='bar')
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("images/top_genres.png")
plt.show()
```

```python
### Top 10 Directors

### Split multiple directors if present
directors = df['director'].str.split(', ', expand=False).explode()

top_directors = directors[directors != 'Unknown'].value_counts().head(10)
print(top_directors)

### Plot
top_directors.plot(kind='bar')
plt.title("Top 10 Directors on Netflix")
plt.xlabel("Director")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("images/top_directors.png")
plt.show()
```

```python
### Top 10 Actors/Cast Members

### Split cast members
cast_members = df['cast'].str.split(', ', expand=False).explode()

top_cast = cast_members[cast_members != 'Unknown'].value_counts().head(10)
print(top_cast)

### Plot
top_cast.plot(kind='bar')
plt.title("Top 10 Most Featured Cast Members on Netflix")
plt.xlabel("Actor / Cast Member")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("images/top_cast.png")
plt.show()
```

```python
### Movie Duration Analysis

### Filter movies only
movies_df = df[df['type'] == 'Movie'].copy()

### Extract numeric duration from strings like "90 min"
movies_df['duration_int'] = movies_df['duration'].str.extract(r'(\d+)')
movies_df['duration_int'] = pd.to_numeric(movies_df['duration_int'], errors='coerce')

### Drop missing duration values
movies_df = movies_df.dropna(subset=['duration_int'])

### Plot movie duration distribution
sns.histplot(movies_df['duration_int'], bins=30, kde=True)
plt.title("Distribution of Movie Durations on Netflix")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("images/movie_duration_distribution.png")
plt.show()

### Summary stats
print("\nMovie Duration Summary:")
print(movies_df['duration_int'].describe())
```

```python
### TV Show Seasons Analysis

### Filter TV Shows only
tv_df = df[df['type'] == 'TV Show'].copy()

### Extract number of seasons from strings like "2 Seasons"
tv_df['seasons'] = tv_df['duration'].str.extract(r'(\d+)')
tv_df['seasons'] = pd.to_numeric(tv_df['seasons'], errors='coerce')

### Drop missing season values
tv_df = tv_df.dropna(subset=['seasons'])

### Plot TV show seasons distribution
sns.countplot(data=tv_df, x='seasons', order=tv_df['seasons'].value_counts().index[:10])
plt.title("Distribution of Number of Seasons in TV Shows")
plt.xlabel("Number of Seasons")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("images/tv_show_seasons.png")
plt.show()

### Summary stats
print("\nTV Show Seasons Summary:")
print(tv_df['seasons'].describe())
```

```python
### Top 10 Content-Producing Years

top_release_years = df['release_year'].value_counts().head(10)
print(top_release_years)

top_release_years.sort_index().plot(kind='bar')
plt.title("Top Release Years by Number of Titles")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("images/top_release_years.png")
plt.show()
```
